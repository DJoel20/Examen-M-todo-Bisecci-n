```python
import numpy as np
from scipy.optimize import bisect
import matplotlib.pyplot as plt

def f(x):
    return x**5 - 6*x**4 + 2*x**3 + 20*x**2 - 27*x + 10

intervals = [(-2, -1), (-1, 0), (0, 1), (1, 2), (2, 3), (3, 4)]

roots = []

for (a, b) in intervals:
    try:
        root = bisect(f, a, b, xtol=2e-12, rtol=np.float64(8.881784197001252e-16), maxiter=100, disp=True)
        roots.append(root)
    except ValueError:
        continue

roots = sorted(set(np.round(roots, decimals=12)))

print("Raices:", roots)

x = np.linspace(-2, 4, 400)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r'$x^5 - 6x^4 + 2x^3 + 20x^2 - 27x + 10$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gr√°fico')
plt.legend()
plt.show()
```
